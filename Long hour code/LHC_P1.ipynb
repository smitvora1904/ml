# TASK-1

```python
data_death = pd.read_csv("time_series_covid19_deaths_global.csv")
data_recovered = pd.read_csv("time_series_covid19_recovered_global.csv")
data_confirmed =   pd.read_csv("time_series_covid19_confirmed_global.csv")

data_death = pd.DataFrame(data_death)
data_recovered = pd.DataFrame(data_recovered)
data_confirmed = pd.DataFrame(data_confirmed)
```

```python
data_death_india = data_death.loc[data_death['Country/Region']=='India']
data_death_india = data_death_india.drop(['Province/State', 'Country/Region','Lat','Long'], axis = 1)
data_death_india.index = ["Deaths"]
```

```python
new_df = data_death_india.transpose()
```

```python
data_recovered_india = data_recovered.loc[data_recovered['Country/Region'] == 'India']
data_recovered_india = data_recovered_india.drop(['Province/State', 'Country/Region','Lat','Long'], axis = 1)
data_recovered_india.index = ["recovered"]
temp = data_recovered_india.transpose()
new_df['recovered'] = temp
```

```python
data_confirmed_india = data_confirmed.loc[data_confirmed['Country/Region'] == 'India']
data_confirmed_india = data_confirmed_india.drop(['Province/State', 'Country/Region','Lat','Long'], axis = 1)
data_confirmed_india.index = ["confirmed"]
temp = data_confirmed_india.transpose()
new_df['confirmed'] = temp
```

```python
new_df
```

```python
plt.plot(new_df.index, new_df['Deaths'], label = "Deaths")
plt.plot(new_df.index, new_df['recovered'], label = "recovered")
plt.plot(new_df.index, new_df['confirmed'], label = "confirmed")
plt.xlabel("Days ")
plt.ylabel("Cases Death,Recovered,Confirmed")
plt.title("India Covid-19 from 22-jan to today")
plt.legend()
plt.show()
```

# Train the ML model with the hypothesis of regression for India and predict the number of  covid cases on 4th May, 2021 (i.e. Today)


```python
plt.plot(new_df.index, new_df['confirmed'], label = "confirmed")
plt.xlabel("Days ")
plt.ylabel("Cases Confirmed")
plt.title("India Covid-19 from 22-jan to today")
plt.legend()
plt.show()
```

```python
data_pred = [i for i in range(1,469)]
df = new_df
```

```python
df.insert(0,"Days",data_pred,True)
```

```python
df
```

```python
X_features = df.iloc[:,0:1]
Y_features = df.iloc[:,-1]
```

```python
from sklearn.linear_model import LinearRegression
prediction_model = LinearRegression().fit(X_features, Y_features)
x = np.array([469]).reshape(-1, 1)
x
```

```python
y_pred = prediction_model.predict(x)
print(int(y_pred))
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 6)
X_poly = poly.fit_transform(X_features)
poly = poly.fit(X_poly, Y_features)
lin2 = LinearRegression()
lin2.fit(X_poly, Y_features)
x11 = np.array([469]).reshape(-1, 1)
x12 = np.array([468]).reshape(-1, 1)
ans11 = int(lin2.predict(poly.fit_transform(x11)))
ans12 = int(lin2.predict(poly.fit_transform(x12)))
print(ans11)  #19925604
print(ans12)  #19925604
```

```python
print("Total Case on 4th May in India : ",(ans11 - ans12))
fin_ind = ans11-ans12
```

```python
data_pred_guj = [i for i in range(1,65)]
df2 = data_india_gujrat
df2.insert(0,"Days",data_pred_guj,True)
print(df2)
```

```python
X_features = df2.iloc[:,0:1]
Y_features = df2.iloc[:,8:9]

print(Y_features)

```

```python
from sklearn.linear_model import LinearRegression
prediction_model = LinearRegression().fit(X_features, Y_features)
x = np.array([65]).reshape(-1, 1)
x

y_pred_gujrat = prediction_model.predict(x)
print(int(y_pred_gujrat))
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 4)
X_poly = poly.fit_transform(X_features)
poly = poly.fit(X_poly, Y_features)
lin2 = LinearRegression()
lin2.fit(X_poly, Y_features)
x = np.array([65]).reshape(-1, 1)
x2 = np.array([64]).reshape(-1, 1)
ans2 =int(lin2.predict(poly.fit_transform(x))) 
ans3 =int(lin2.predict(poly.fit_transform(x2))) 
print(ans2)
print(ans3)
```

```python
print("Total Case in gujrat on 4th May : ",(ans2-ans3))
fin_guj = ans2-ans3
```
